```python
import requests
import json

keywords_list = ["coffee", "vehicle access system", "wind turbine"]

for keyword in keywords_list:
    query = keyword.replace(" ", "%2B")
    url = f"https://patents.google.com/xhr/query?url=q%3D({query})"

# Anfrage senden und Antwort erhalten
response = requests.get(url)

# Überprüfen, ob die Anfrage erfolgreich war (Status Code 200)
if response.status_code == 200:
    # JSON-Daten aus der Antwort extrahieren
    data = response.json()

    print("JSON-Daten erfolgreich geladen.")
else:
    print(f"Fehler beim Herunterladen der Daten. Statuscode: {response.status_code}")
```
